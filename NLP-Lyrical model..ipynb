```python
from lyricsgenius import Genius
import time
import re
from requests.exceptions import Timeout
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tqdm import tqdm

genius = Genius("TPtNd1DXHEgMmquLES59h5OAwrWf_0pWT8fxyR4peKDfoPt7WVPpHCAoxI8H65Mf", remove_section_headers=True, skip_non_songs=True, excluded_terms=["(Remix)", "(Live)"], timeout=15, retries=3)
analyser = SentimentIntensityAnalyzer()

def fetch_song_lyrics(title, artist, max_retries=3, delay=2, timeout=10):
    for i in range(max_retries):
        try:
            song = genius.search_song(title, artist)
            if song:  # Check if the song was found
                lyrics = song.lyrics
                return clean_lyrics(lyrics)
        except Exception as e:
            if i < max_retries - 1:  # i is zero indexed
                time.sleep(delay)  # Wait for the specified delay before retrying
            else:
                print(f"Failed to retrieve song after {max_retries} attempts.")
                return None



def clean_lyrics(lyrics):
    # Removing contributor/translation information
    lyrics = re.sub(r"^[^\n]*Lyrics", "", lyrics)

    # Removing [Chorus], [Verse] etc.
    lyrics = re.sub(r"\[.*\]\n", "", lyrics)
    
    # Replacing multiple newlines with a single newline
    lyrics = re.sub(r"\n{2,}", "\n", lyrics)

    # Removing leading/trailing whitespace
    lyrics = lyrics.strip()
    
    return lyrics

# Initialize lists for results
user_sentiments = []
user_sentiment_scores = []

# Loop through the users
for user_name, user_data in tqdm(df.iterrows(),total=df.shape[0], desc="Processing users"):
    user_total_sentiment = 0
    song_count = 0
    
    for col in user_data.index:
        if not isinstance(user_data[col], str) or ' by ' not in user_data[col]:  # Skip entries without ' by ' or non-string entries
            continue

        retries = 3
        for _ in range(retries):
            try:
                song_name, artist_name = user_data[col].split(' by ')
                
                # Search for the song lyrics using the Genius API
                song = genius.search_song(song_name.strip(), artist_name.strip()) # strip() to remove extra spaces
                if song:  # Check if the song was found
                    lyrics = clean_lyrics(song.lyrics)
                    sentiment_score = analyser.polarity_scores(lyrics)
                    user_total_sentiment += sentiment_score['compound']
                    song_count += 1
                
                break  # If successful, break out of the retry loop
                
            except Timeout:
                print(f"Attempt {_+1} failed. Retrying...")
                if _ < retries - 1:  # No need to sleep after the last attempt
                    time.sleep(2)  # Wait for 2 seconds before retrying
            except Exception as e:
                print(f"An error occurred: {e}")
                break  # If it's an error other than Timeout, break out of the retry loop

    # Calculate average sentiment for the user
    if song_count > 0:
        average_sentiment = user_total_sentiment / song_count
    else:
        average_sentiment = 0
    
    if average_sentiment >= 0.05:
        sentiment = 'Positive'
    elif average_sentiment > -0.05 and average_sentiment < 0.05:
        sentiment = 'Neutral'
    elif average_sentiment <= -0.05:
        sentiment = 'Negative'
    
    user_sentiments.append(sentiment)
    user_sentiment_scores.append(abs(average_sentiment) * 100)

# Add user sentiments to the dataframe
df['User_Sentiment'] = user_sentiments
df['User_Sentiment_Score'] = user_sentiment_scores
```

```python

import pandas as pd
import requests
import time
from tqdm import tqdm  # Import the tqdm library for progress tracking

# function to fetch the Wikipedia introduction
def get_wikipedia_description(topic):
    base_url = "https://en.wikipedia.org/w/api.php"
    
    # variations: capitalize the first letter and the entire string
    variations = [topic, topic.capitalize(), topic.upper()]
    for var in variations:
        params = {
            "action": "query",
            "titles": var,
            "prop": "extracts",
            "exintro": True,  # Get only the introduction/summary
            "format": "json"
        }
        response = requests.get(base_url, params=params)
        data = response.json()
        pages = data["query"]["pages"]
        for _, page in pages.items():
            extract = page.get("extract", None)
            if extract:
                return extract

        time.sleep(2)  # Introducing a 2-second delay between requests

    # Return None if no description is found for all variations
    return None

# Load the dataset
hobby_data = pd.read_csv("hobbylist.csv")

# Fetch the Wikipedia introductions for the hobbies in the dataset
descriptions = []
for hobby in tqdm(hobby_data['Hobby-name'], desc="Fetching Descriptions"):
    descriptions.append(get_wikipedia_description(hobby))
hobby_data['Description'] = descriptions

# Identify hobbies with missing descriptions
hobby_data['Not Found'] = hobby_data['Description'].isnull()

# Fetch the Wikipedia introductions for the explicitly mentioned activities
additional_hobbies = ["Swimming", "Reading", "Chess"]
additional_descriptions = [get_wikipedia_description(hobby) for hobby in additional_hobbies]
additional_not_found = [desc is None for desc in additional_descriptions]
additional_data = pd.DataFrame({
    'Hobby-name': additional_hobbies,
    'Description': additional_descriptions,
    'Not Found': additional_not_found
})

# Combine the data
combined_data = pd.concat([hobby_data, additional_data], ignore_index=True)

# Save the combined data to a .csv file
combined_data.to_csv("test_hobbies_with_descriptions.csv", index=False)
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to standardize hobby names
#def standardize_hobby_name(name):
    #return str(name).strip().lower()

# Load datasets
hobbies_df = pd.read_csv("hobbies_with_cleaned_descriptions.csv")
users_df = pd.read_csv("User_Sentiment_Scores_Lyrics.csv")

# Standardize hobby names
hobbies_df["Hobby-name"] = hobbies_df["Hobby-name"].apply(standardize_hobby_name)
users_df["Hobby"] = users_df["Hobby"].apply(standardize_hobby_name)

# Compute sentiment scores for hobby descriptions
hobbies_df["Sentiment_Score"] = hobbies_df["Description"].apply(lambda x: analyzer.polarity_scores(str(x))['compound'] if pd.notna(x) and x != "Not Found" else 0)

# Update user sentiment score
def aggregate_scores(user_row):
    filtered_hobbies = hobbies_df[hobbies_df["Hobby-name"] == user_row["Hobby"]]["Sentiment_Score"]
    
    # Check if the filtered DataFrame has any rows
    if not filtered_hobbies.empty:
        hobby_score = filtered_hobbies.iloc[0]
    else:
        hobby_score = 0  # Default value for unmatched hobbies
    
    playlist_score = user_row["Sentiment_Score_From_Lyrics"]

    # Weighted average: 70% playlist_score, 30% hobby_score
    return 0.7 * playlist_score + 0.3 * hobby_score

# Rename column
users_df.rename(columns={"User_Sentiment_Score": "Sentiment_Score_From_Lyrics"}, inplace=True)

# Drop duplicates in hobbies_df based on the "Hobby-name" column
hobbies_df = hobbies_df.drop_duplicates(subset="Hobby-name")

# Now, proceed with the mapping
users_df["Sentiment_Score_From_Description"] = users_df["Hobby"].map(hobbies_df.set_index("Hobby-name")["Sentiment_Score"]).fillna(0)

# Compute the aggregated user sentiment score
users_df["Aggregated_User_Sentiment_Score"] = users_df.apply(aggregate_scores, axis=1)

# Classify the sentiment based on the range of scores
def normalize_to_vader_scale(score, score_series):
    current_min = score_series.min()
    current_max = score_series.max()
    return 2 * ((score - current_min) / (current_max - current_min)) - 1

def classify_sentiment(score, score_series):
    normalized_score = normalize_to_vader_scale(score, score_series)
    if normalized_score > 0.05:
        return 'Positive'
    elif -0.05 <= normalized_score <= 0.05:
        return 'Neutral'
    else:
        return 'Negative'

users_df["Aggregated_Sentiment"] = users_df["Aggregated_User_Sentiment_Score"].apply(lambda score: classify_sentiment(score, users_df["Aggregated_User_Sentiment_Score"]))

# Drop the User_Sentiment_Score column since it's not needed
users_df.drop(columns=["User_Sentiment"], inplace=True, errors='ignore')  # Added error handling in case column doesn't exist

cols = users_df.columns.tolist()
cols.insert(2, cols.pop(cols.index('Aggregated_Sentiment')))
cols.insert(3, cols.pop(cols.index('Aggregated_User_Sentiment_Score')))
cols.insert(5, cols.pop(cols.index('Sentiment_Score_From_Description')))
users_df = users_df[cols]

# Save the updated data
hobbies_df.to_csv("updated_hobbies.csv", index=True)
users_df.to_csv("updated_users_data.csv", index=True)
```

```python
import numpy as np

# Define sentiment ranges
buckets = {
    "Very Negative": (-1, -0.6),
    "Negative": (-0.6, -0.2),
    "Neutral": (-0.2, 0.2),
    "Positive": (0.2, 0.6),
    "Very Positive": (0.6, 1)
}

def determine_bucket(score):
    for bucket_name, (lower, upper) in buckets.items():
        if lower <= score <= upper:
            return bucket_name
    return "Neutral"

# Assign each Hobby to a sentiment bucket
hobbies_df["Bucket"] = hobbies_df["Sentiment_Score"].apply(determine_bucket)

def predict_hobby(user_score, hobbies_df):
    user_bucket = determine_bucket(user_score)
    possible_hobbies = hobbies_df[hobbies_df["Bucket"] == user_bucket]
    
    # If there are no hobbies in this bucket, default to Neutral
    if possible_hobbies.empty:
        user_bucket = "Neutral"
        possible_hobbies = hobbies_df[hobbies_df["Bucket"] == user_bucket]
    
    # Recommend a random hobby from the possible hobbies
    return np.random.choice(possible_hobbies["Hobby-name"])

users_df["Predicted_Hobby"] = users_df["Aggregated_User_Sentiment_Score"].apply(lambda x: predict_hobby(x, hobbies_df))

if "Predicted_Activity" in users_df.columns:
    users_df.drop(columns=["Predicted_Activity"], inplace=True)


cols = users_df.columns.tolist()
cols.insert(2, cols.pop(cols.index('Predicted_Hobby')))
users_df = users_df[cols]
```

```python
import numpy as np
import pandas as pd
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def load_data():
    hobbies_data = pd.read_csv("updated_hobbies.csv", index_col=0)
    users_data = pd.read_csv("updated_users_data.csv", index_col=0)
    songs_data = pd.read_csv("sampled_songs_lyrics.csv")
    return hobbies_data, users_data, songs_data
```

```python
    
def clean_lyrics(lyrics):
    return ' '.join(re.sub(r'[^a-zA-Z\s]', '', re.sub(r'\[.*?\]', '', lyrics)).lower().split())
```

```python
def compute_track_sentiment(lyrics, analyzer):
    return analyzer.polarity_scores(clean_lyrics(lyrics))['compound']
```

```python
def get_hobby_sentiment(hobby, hobbies_df):
    matching_hobbies = hobbies_df[hobbies_df['Hobby-name'].str.contains(hobby, case=False)]
    return matching_hobbies['Sentiment_Score'].iloc[0] if not matching_hobbies.empty else np.nan
```

```python
def recommend_tracks_for_user(username, users_df, df,hobbies_df):
    if username not in users_df.index:
        return []
    
    user_hobby = users_df.loc[username, 'Hobby']
    user_hobby_sentiment = get_hobby_sentiment(user_hobby, hobbies_df)  # Look up in hobbies_df
    
    df_copy = df.copy()
    df_copy['Aggregate_Sentiment'] = 0.7 * df_copy['Track_Sentiment'] + 0.3 * user_hobby_sentiment
    
    
    
    recommended_tracks = df_copy[np.isclose(df_copy['Aggregate_Sentiment'], user_hobby_sentiment, atol=0.1)]
    if len(recommended_tracks) < 5:
        recommended_tracks = df_copy[np.isclose(df_copy['Aggregate_Sentiment'], user_hobby_sentiment, atol=0.2)]
    
    recommended_tracks['Track_Artist'] = recommended_tracks['title'] + " by " + recommended_tracks['artist']
    
    return recommended_tracks.sample(min(5, len(recommended_tracks)))['Track_Artist'].tolist()
```

```python
if __name__ == "__main__":
    # Initialize the VADER analyzer and load data
    analyzer = SentimentIntensityAnalyzer()
    hobbies_df, users_df, df = load_data()
    
    # Compute sentiment scores for tracks and hobbies
    df['Track_Sentiment'] = df['lyrics'].apply(lambda x: compute_track_sentiment(x, analyzer))

    # Make recommendations for each user
    users_df['Recommended_Tracks'] = users_df.index.map(lambda x: recommend_tracks_for_user(x, users_df, df,hobbies_df))
    
   
```
